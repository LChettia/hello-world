```python
for pdate in pdr:
    output_index = df_output.loc[df_output['unit_balance']>0, 'production_date']==pdate
    output_index_list = output_index[output_index].index
    if output_index_list.empty:
        print('Production date: {} No output projection' .format(pdate))
        
    else:
        for o_index in output_index_list:
            intake_index = df_intake.sort_values(by='plot_rank').loc[df_intake['unit_balance']>0, 'tat_start_date']>=pdate
            intake_index_list = intake_index[intake_index].index
            
            if intake_index_list.empty:
                print('Production date: {} No intake projection' .format(pdate))
            else:
                for i_index in intake_index_list:
                    in_units == df_intake.loc[i_index, 'unit_balance']
                    out_units == df_output.loc[o_index, 'unit_balance']
                    plot_unit = min(in_units, out_units)
                    
                    if plot_unit >=0:
                        print('production date: {} output_index: {} intake_index: {} stone_unit: {}' .format(pdate, o_index, i_index, units) )
                        prod_plan = prod_plan.append({'production_date': pdate, 'df_output_index': o_index, 'df_intake_index': i_index, 'stone_units_plot': units}, ignore_index=True)
                        df_intake.loc[i_index, 'unit_balance'] = df_intake.loc[i_index, 'unit_balance'] - plot_unit
                        df_output.loc[o_index, 'unit_balance'] = df_intake.loc[o_index, 'unit_balance'] - plot_unit
                    else:
                        break
                    
            
```
